```python
import pandas as pd

train_data = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
test_data.head()
```

```python
train_data.columns
```

```python
# LETS START
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error


X_train = train_data.drop('label', axis=1).values
y_train = train_data['label'].values
X_test = test_data.values


# Normalize the data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Split the data
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)



# MAKE MODEL
model = DecisionTreeClassifier(random_state=62, max_depth=20)

# Train the model on the split training data
model.fit(X_train_split, y_train_split)

# Predict on the validation set
pred = model.predict(X_val)

# Check MAE
mean_absolute_error(y_val, pred)

# Train the model on the full training data
model.fit(X_train, y_train)

# Predict on the test data
predictions = model.predict(X_test)

# SUBMISSION
submission = pd.DataFrame({'ImageId': np.arange(1, len(predictions) + 1), 'Label': predictions})

# Save the submission file
submission.to_csv('submission.csv', index=False)
print("Submission successfully received!")
```
